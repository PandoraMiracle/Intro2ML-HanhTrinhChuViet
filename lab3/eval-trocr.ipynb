```python
import os, json, random, argparse, unicodedata
from typing import List, Dict, Any, Tuple

import torch
from jiwer import wer, cer
```

```python
# ========= (1) Text utils =========
def normalize_text(s: str) -> str:
    # Giữ dấu tiếng Việt ổn định (NFC), normalize whitespace nhẹ
    s = unicodedata.normalize("NFC", s or "")
    s = s.replace("\r\n", "\n").replace("\r", "\n")
    s = " ".join(s.split())
    return s.strip()

def set_seed(seed: int):
    random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
```

```python
import json
import os

def load_split(split_path):
    with open(split_path, "r", encoding="utf-8") as f:
        sp = json.load(f)
    return sp["train"], sp["val"]

_, val_pairs = load_split("/kaggle/working/split.json")
print("val size:", len(val_pairs))
# Example item:", val_pairs[0] if val_pairs else None
```

```python
import random

def run_eval_generic(val_pairs, predict_one_fn, max_eval=200):
    samples = random.sample(val_pairs, min(max_eval, len(val_pairs)))
    y_true, y_pred = [], []
    for s in samples:
        gt = normalize_text(s["text"])
        pr = normalize_text(predict_one_fn(s["image_path"]))
        y_true.append(gt); y_pred.append(pr)
    return {"cer": float(cer(y_true, y_pred)), "wer": float(wer(y_true, y_pred)), "n": len(samples)}
```

## TrOCR handwritten

```python
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"
os.environ["TRANSFORMERS_NO_FLAX"] = "1"

import torch
from transformers import TrOCRProcessor, VisionEncoderDecoderModel

trocr_id = "microsoft/trocr-base-handwritten"
processor = TrOCRProcessor.from_pretrained(trocr_id)
trocr = VisionEncoderDecoderModel.from_pretrained(trocr_id).to("cuda" if torch.cuda.is_available() else "cpu")
trocr.eval()

print("TrOCR OK")

from PIL import Image

@torch.inference_mode()
def trocr_predict_one(image_path: str) -> str:
    img = Image.open(image_path).convert("RGB")
    pixel_values = processor(images=img, return_tensors="pt").pixel_values.to(trocr.device)
    generated_ids = trocr.generate(pixel_values, max_new_tokens=128)
    text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return text.strip()
```

```python
# TrOCR

trocr_metrics = run_eval_generic(val_pairs, trocr_predict_one, max_eval=200)
print("TrOCR:", trocr_metrics)
```
